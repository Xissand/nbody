```python
import numpy as np
import pandas as pd
```

```python
index = 0
dt = 0.00001
dfx = pd.DataFrame(columns=['time','diff_x'])
dfv = pd.DataFrame(columns=['time','diff_v'])
with open("../part1.xyz", "r") as file1:
    with open("../part2.xyz", "r") as file2:
        while(1):
            index+=1
            try:
                n = int(file1.readline())
                file2.readline()
            except:
                break
            file1.readline()
            file2.readline()
            
            coord1 = []
            coord2 = []
            difx = np.zeros(n)
            difv = np.zeros(n)
            
            for i in range(0,n):
                line1 = file1.readline()
                coord1 = [float(j) for j in line1.split(" ")]
                line2 = file2.readline()
                coord2 = [float(j) for j in line2.split(" ")]
                
                difx[i] += (coord1[0]-coord2[0])**2
                difx[i] += (coord1[1]-coord2[1])**2
                difx[i] += (coord1[2]-coord2[2])**2
                
                difv[i] += (coord1[3]-coord2[3])**2
                difv[i] += (coord1[4]-coord2[4])**2
                difv[i] += (coord1[5]-coord2[5])**2
            
            dfx.loc[index]=[index*dt,np.mean(difx)]
            
            dfv.loc[index]=[index*dt,np.mean(difv)]

print(dfx)
print(dfv)
```
